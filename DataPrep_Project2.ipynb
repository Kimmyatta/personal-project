**Correlation Heatmap**

This is a specific form of heat map that depicts the correlation coefficients between numerous variables in a dataset.  Correlation heat maps are commonly used in statistical and data analysis to demonstrate how closely variables are related to one another.  For example:
 Values near 1 (or -1) imply a significant positive (or negative) association.
 Values approaching zero suggest no association.  A correlation heat map's colors typically span from one shade (e.g., dark blue) to another (e.g., blazing red), reflecting the intensity and direction of the correlation.

 So, whereas a basic heat map can depict any type of data intensity, a correlation heat map is designed primarily to visualize correlations between variables.



```python
co_mtx = df.corr(numeric_only=True)

# Print correlation matrix
print(co_mtx)

# Plot correlation heatmap
sns.heatmap(co_mtx, cmap="YlGnBu", annot=True)

# Display heatmap
plt.show()
```

**Pairplot**

A pairplot is a data visualization that plots pairwise relationships between variables in a dataset. It allows you to visualize the relationships between pairs of variables, making it easier to understand the data's structure and correlations. Here we are trying to print the patterns that vary between males and females in the dataset.

```python
import seaborn as sns # Make sure seaborn is imported with the alias 'sns'
import matplotlib.pyplot as plt

df = sns.load_dataset('tips') # Use the alias 'sns' here
# pairplot with hue sex
sns.pairplot(df, hue='sex') # Use the alias 'sns' here
# to show
plt.show()
```

**Violin plot**

 A violin plot performs a similar function as the whisker or box plot.  As it displays many quantitative data points across one or more categorical variables.  It can be an effective and appealing approach to display several data points at different scales.  A "wide-form" Data Frame helps to keep track of each numerical column that can be plotted on the graph.  You can use NumPy or Python objects, but pandas objects are preferred because the associated names will be utilized to annotate the axes. This code creates a violin plot to represent the distribution of 100 randomly generated data points. The violin plot provides insight into the data's density, median, and overall form, which aids in understanding the dataset's features..




```python
import matplotlib.pyplot as plt
import numpy as np

# Generate random data
np.random.seed(1)
data = np.random.randn(100)

# Create a violin plot
plt.figure()
plt.violinplot(data, showmedians=True)

# Set plot labels and title
plt.xlabel('Variable')
plt.ylabel('Value')
plt.title('Univariate Violin Plot')

# Show the plot
plt.show()
```
